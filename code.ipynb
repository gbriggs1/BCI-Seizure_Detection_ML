```python
import pandas as pd
import numpy as np
import math
#neural networks, support vector machines, random forests anda KNN classifier. 
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
from sklearn import svm
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.naive_bayes import GaussianNB
```

- The response variable is y in column 179
- All subjects falling in classes 2, 3, 4, and 5 are subjects who did not have epileptic seizure
- 1- Recording of seizure activity
- The Explanatory variables X1, X2, ..., X178

```python
df = pd.read_csv("data.csv")
df
```

```python
df = df.drop(columns=['Unnamed: 0'])
df.loc[df['y'] > 1, 'y'] = 0
df
```

```python
y = df['y']
x = df
x = x.drop(columns=['y'])
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=0)
```

```python
rfcl = RandomForestClassifier(max_depth=5, random_state=1)
rfcl.fit(x_train, y_train)
rfcl.score(x_test,y_test)
```

Random Forest Classifier has an accuracy of 94.96%

```python
# clf = svm.SVC(kernel='precomputed')
clf = svm.SVC(kernel='rbf')
clf.fit(x_train, y_train)
clf.score(x_test, y_test)
```

Support Vector Machine Classifier has an accuracy of 97.6% 

```python
nncl = MLPClassifier(solver='lbfgs', hidden_layer_sizes=(10,), random_state=1)
nncl.fit(x_train, y_train)
nncl.score(x_test, y_test)
```

Neural network has an accuracy of 96.21%

```python
neigh = KNeighborsClassifier(n_neighbors=3)
neigh.fit(x_train, y_train)
neigh.score(x_test, y_test)
```

KNN has an accuracy of 92.87

```python
gnb = GaussianNB()
gnb.fit(x_train, y_train)
gnb.score(x_test, y_test)
```
